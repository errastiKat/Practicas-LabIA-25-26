```python
import numpy as np
import pandas as pd
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

Variables:

**Antecedentes (Entradas)**
*   Antig√ºedad del cliente
*   Frecuencia de compra
*   Gasto media por compra
*   D√≠as desde la √∫ltima compra
*   Variedad de productos comprados
*   Gasto total
*   Productos comprados >1 vez


**Consecuentes (Salidas)**
*   Cliente en riesgo de abandono
*   Comprador ocasional
*   Explorador curioso

*   Cliente de producto unico





```python
# Antecedentes
antiguedad = ctrl.Antecedent(np.arange(0, 501, 1), 'antiguedad') # Dias desde primera compra (hemos mirado en el csv, y es en un a√±o de periodo (365 dias))
frecuencia = ctrl.Antecedent(np.arange(0, 31, 0.2), 'frecuencia') # Numero compras por mes
gasto_medio = ctrl.Antecedent(np.arange(0, 25001, 3), 'gasto_medio') # Precio medio por compra
dias_ultima_compra = ctrl.Antecedent(np.arange(0, 501, 1), 'dias_ultima_compra') # Dias desde ultima compra (hemos mirado en el csv, y es en un a√±o de periodo (365 dias))
variedad = ctrl.Antecedent(np.arange(0, 2251, 1), 'variedad') # Cantidad productos distintos (hemos mirado que el maximo en el csv en 2156 -- redondeamos a 2250)
total_gasto = ctrl.Antecedent(np.arange(0, 300001, 15), 'total_gasto') # gasto total
productos_repetidos = ctrl.Antecedent(np.arange(0, 1400, 1), 'productos_repetidos') # productos comprados >1 vez

# Universo com√∫n para todas las salidas
universo = np.arange(0, 101, 1)

# --- Definir cada categor√≠a como Consecuencia independiente ---
riesgo_abandono = ctrl.Consequent(universo, 'riesgo_abandono')
comprador_ocasional = ctrl.Consequent(universo, 'comprador_ocasional')
explorador = ctrl.Consequent(universo, 'explorador')
toda_vida = ctrl.Consequent(universo, 'toda_vida')
producto_unico = ctrl.Consequent(universo, 'producto_unico')
```

# **Funciones de pertenencia:** antecedentes

```python
# Antiguedad
antiguedad['baja'] = fuzz.trimf(antiguedad.universe, [0, 0, 25])
antiguedad['media'] = fuzz.trimf(antiguedad.universe, [0, 25, 150])
antiguedad['alta'] = fuzz.trapmf(antiguedad.universe, [25, 150, 500, 500])

# Frecuencia
frecuencia['baja'] = fuzz.trimf(frecuencia.universe, [0, 0, 0.75])
frecuencia['media'] = fuzz.trimf(frecuencia.universe, [0, 0.75, 1.25])
frecuencia['alta'] = fuzz.trapmf(frecuencia.universe, [0.75, 1.25, 31, 31])

# Gasto medio
gasto_medio['bajo'] = fuzz.trimf(gasto_medio.universe, [0, 0, 220])
gasto_medio['medio'] = fuzz.trimf(gasto_medio.universe, [0, 220, 375])
gasto_medio['alto'] = fuzz.trapmf(gasto_medio.universe, [220, 375, 25000, 25000])

# Ultima compra
dias_ultima_compra['reciente'] = fuzz.trimf(dias_ultima_compra.universe, [0, 0, 20])
dias_ultima_compra['intermedia'] = fuzz.trimf(dias_ultima_compra.universe, [0, 20, 50])
dias_ultima_compra['lejana'] = fuzz.trapmf(dias_ultima_compra.universe, [20, 50, 500, 500])

# Variedad
variedad['poca'] = fuzz.trimf(variedad.universe, [0, 0, 40])
variedad['moderada'] = fuzz.trimf(variedad.universe, [0, 40, 70])
variedad['muchas'] = fuzz.trapmf(variedad.universe, [40, 70, 2250, 2250])

# Total gasto
total_gasto['bajo'] = fuzz.trimf(total_gasto.universe, [0, 0, 400])
total_gasto['medio'] = fuzz.trimf(total_gasto.universe, [0, 400, 1300])
total_gasto['alto'] = fuzz.trapmf(total_gasto.universe, [400, 1300, 3000000, 3000000])

# Productos repetidos
productos_repetidos['pocos'] = fuzz.trimf(productos_repetidos.universe, [0, 0, 2])
productos_repetidos['moderados'] = fuzz.trimf(productos_repetidos.universe, [0, 2, 15])
productos_repetidos['muchos'] = fuzz.trapmf(productos_repetidos.universe, [2, 15, 1400, 1400])
```

# **Cargar y limpiar** dataset retail.csv

```python
datos_retail = pd.read_csv("retail.csv", parse_dates=["InvoiceDate"])

# Filtrar transacciones validas (cantidad y precio positivos, evitar devoluciones, eliminar
# casos anomalos)
datos_retail = datos_retail[(datos_retail["Quantity"] > 0) & (datos_retail["UnitPrice"] > 0) & (datos_retail["Quantity"] < 500)] # Eliminar los que devuelven productos, pues nos interesa solo lo que compran

datos_retail['Gasto'] = datos_retail['Quantity'] * datos_retail['UnitPrice']
```

# **Agrupar datos por cliente**


```python
clientes = datos_retail.groupby('CustomerID').agg(
    primera_compra=('InvoiceDate', 'min'), # Fecha de primera compra de ese cliente
    ultima_compra=('InvoiceDate', 'max'), # Fecha de ultima compra de ese cliente
    total_compras=('InvoiceNo', 'nunique'), # Numero de compras totales de mismo cliente
    total_gasto=('Gasto', 'sum'), # Suma de gastos de lineas con mismo CustomerID
    variedad=('StockCode', 'nunique') # Cantidad de productos distintos comprados con mismo CustomerID
).reset_index()


# Fecha de referencia (ultima fecha del dataset)
clientes['primera_compra'] = pd.to_datetime(clientes['primera_compra'])
clientes['ultima_compra'] = pd.to_datetime(clientes['ultima_compra'])
fecha_actual = clientes['ultima_compra'].max()

# Calcular indicadores
clientes['gasto_medio'] = clientes['total_gasto'] / clientes['total_compras'] # gasto medio por compra
clientes['antiguedad'] = (clientes['ultima_compra'] - clientes['primera_compra']).dt.days
clientes['meses_activos'] = ((clientes['ultima_compra'].dt.year - clientes['primera_compra'].dt.year) * 12 + (clientes['ultima_compra'].dt.month - clientes['primera_compra'].dt.month))
clientes['frecuencia'] = clientes['total_compras'] / (clientes['meses_activos'].replace(0, 1)) # evitar divisi√≥n por 0
clientes['dias_ultima_compra'] = (fecha_actual - clientes['ultima_compra']).dt.days

repetidos = datos_retail.groupby(['CustomerID', 'StockCode']).size().reset_index(name='conteo')
repetidos = repetidos[repetidos['conteo'] > 1].groupby('CustomerID').size().reset_index(name='productos_repetidos')
clientes = clientes.merge(repetidos, on='CustomerID', how='left')
clientes['productos_repetidos'] = clientes['productos_repetidos'].fillna(0)

# Columnas con los indicadores (antecedentes) para mostrar que realmente se han calculado
display(clientes[['CustomerID', 'antiguedad', 'frecuencia', 'gasto_medio', 'variedad',
                  'dias_ultima_compra', 'total_gasto', 'productos_repetidos']].head(10))
```

# **Reglas difusas**


```python
reglas = []

# 1) Riesgo de abandono
r1 = ctrl.Rule(
    (antiguedad['alta'] & frecuencia['baja']) |
    dias_ultima_compra['lejana'],
    riesgo_abandono['alto'])
r2 = ctrl.Rule(
    (antiguedad['media'] & frecuencia['baja']) |
    (dias_ultima_compra['intermedia'] & frecuencia['baja']) |
    frecuencia['media'],
    riesgo_abandono['medio']
)
r3 = ctrl.Rule(
    dias_ultima_compra['reciente'] |
    frecuencia['alta'] |
    antiguedad['baja'],
    riesgo_abandono['bajo']
)

# 2) Comprador ocasional
r4 = ctrl.Rule(frecuencia['baja'] & gasto_medio['bajo'], comprador_ocasional['alto'])
r5 = ctrl.Rule(
    (frecuencia['baja'] & gasto_medio['medio']) |
    (frecuencia['media'] & gasto_medio['bajo']),
    comprador_ocasional['medio']
)
r6 = ctrl.Rule(frecuencia['alta'] | gasto_medio['alto'], comprador_ocasional['bajo'])

# 3) Explorador curioso
r7 = ctrl.Rule(
    variedad['muchas'] &
    (gasto_medio['bajo'] | gasto_medio['medio']),
    explorador['alto'])
r8 = ctrl.Rule(
    (variedad['moderada'] & frecuencia['media']) |
    (total_gasto['medio'] & productos_repetidos['pocos']),
    explorador['medio']
)
r9 = ctrl.Rule(variedad['poca'] | gasto_medio['alto'], explorador['bajo'])

# 4) Cliente de toda la vida
r10 = ctrl.Rule(
    (antiguedad['alta'] & (frecuencia['media'] | frecuencia['alta'])) |
    (antiguedad['alta'] & productos_repetidos['muchos']),
    toda_vida['alto']
)
r11 = ctrl.Rule(
    antiguedad['media'] &
    (frecuencia['media'] | frecuencia['alta']),
    toda_vida['medio'])
r12 = ctrl.Rule(antiguedad['baja'] | frecuencia['baja'], toda_vida['bajo'])

# 5) Cliente de producto √∫nico
r13 = ctrl.Rule(
    (variedad['poca'] & productos_repetidos['muchos']) |
    (variedad['poca'] & gasto_medio['alto']),
    producto_unico['alto']
)
r14 = ctrl.Rule(variedad['poca'] & productos_repetidos['moderados'], producto_unico['medio'])
r15 = ctrl.Rule(
    variedad['moderada'] |
    variedad['muchas'] |
    productos_repetidos['pocos'],
    producto_unico['bajo']
)
```

# **Creacion y simulacion** del sistema difuso


```python
# Lista de todas las reglas
reglas = [ r1,r2,r3, r5,r6,r7, r8,r9,r10,r11, r12,r13,r14, r15 ]

# Crear sistema de control
sistema_clientes = ctrl.ControlSystem(reglas)

# Preparacion para simulaci√≥n
simulacion = ctrl.ControlSystemSimulation(sistema_clientes)

# Crear columnas para guardar los resultados
clientes['riesgo_abandono_val'] = 0.0
clientes['comprador_ocasional_val'] = 0.0
clientes['explorador_val'] = 0.0
clientes['toda_vida_val'] = 0.0
clientes['producto_unico_val'] = 0.0

# Ejecutar simulacion para cada cliente
for idx, row in clientes.iterrows():
    simulacion.input['antiguedad'] = row['antiguedad']
    simulacion.input['frecuencia'] = row['frecuencia']
    simulacion.input['gasto_medio'] = row['gasto_medio']
    simulacion.input['dias_ultima_compra'] = row['dias_ultima_compra']
    simulacion.input['variedad'] = row['variedad']
    simulacion.input['total_gasto'] = row['total_gasto']
    simulacion.input['productos_repetidos'] = row['productos_repetidos']

    simulacion.compute()  # Ejecutar la simulacion

    # Guardar los resultados
    clientes.at[idx, 'riesgo_abandono_val'] = simulacion.output.get('riesgo_abandono', 0.0)
    clientes.at[idx, 'comprador_ocasional_val'] = simulacion.output.get('comprador_ocasional', 0.0)
    clientes.at[idx, 'explorador_val'] = simulacion.output.get('explorador', 0.0)
    clientes.at[idx, 'toda_vida_val'] = simulacion.output.get('toda_vida', 0.0)
    clientes.at[idx, 'producto_unico_val'] = simulacion.output.get('producto_unico', 0.0)
```

# **Estadisticas:** porcentaje de pertenencia extrema y percentiles


```python
categorias = ['riesgo_abandono_val', 'comprador_ocasional_val', 'explorador_val', 'toda_vida_val', 'producto_unico_val']

# Tabla 1: Porcentajes de pertenencia extrema
tabla_extremos = []

for cat in categorias:
    total_clientes = len(clientes)
    pct_lt_025 = (clientes[cat] < 25).sum() / total_clientes * 100
    pct_gt_075 = (clientes[cat] > 75).sum() / total_clientes * 100
    tabla_extremos.append([cat.replace('_val',''), round(pct_lt_025,2), round(pct_gt_075,2)])

df_extremos = pd.DataFrame(tabla_extremos, columns=["Categor√≠a", "% <0.25", "% >0.75"])

# Tabla 2: Percentiles 20/40/60/80
tabla_percentiles = []
percentiles = [20, 40, 60, 80]

for cat in categorias:
    p_values = np.percentile(clientes[cat], percentiles)
    tabla_percentiles.append([cat.replace('_val','')] + list(np.round(p_values,2)))

df_percentiles = pd.DataFrame(tabla_percentiles, columns=["Categor√≠a", "P20", "P40", "P60", "P80"])

# --- Mostrar las tablas ---
print("Tabla 1: Porcentajes de pertenencia extrema (<0.25 y >0.75)\n")
print("\t(Indican el porcentaje de cliente que tienen una pertenencia muy baja (menos de 0.25 o 25%) o muy alta (mas de 0.75 o 75%) en cada categoria)\n")
display(df_extremos)

print("\nTabla 2: Percentiles 20/40/60/80 de cada categor√≠a\n")
print("\t(Muestran que, si ordenamos la lista de clientes segun su % de pertenencia a una categoria, el cliente que se encuentra en el percentil x tiene un valor de pertenencia z, y el x% de los demas clientes tiene un valor igual o menor que ese valor de pertenencia z)\n")
display(df_percentiles)
```
